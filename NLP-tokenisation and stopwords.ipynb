### NLP Tokenisation

Tokenisation is to break the complex sentence into words.<br/>
Understand the importance of each words with respect to the sentence<br/>
Produce a structural description on an Input sentence



#### Import libraries

```python
import os
import nltk
import nltk.corpus
```

```python
from nltk.tokenize import word_tokenize
```

```python
word ="""A newspaper is a periodical publication containing written information about current events and is often typed in black ink with a white or gray background.

Newspapers can cover a wide variety of fields such as politics, business, sports and art, and often include materials such as opinion columns, weather forecasts, reviews of local services, obituaries, birth notices, crosswords, editorial cartoons, comic strips, and advice columns.

Most newspapers are businesses, and they pay their expenses with a mixture of subscription revenue, newsstand sales, and advertising revenue. The journalism organizations that publish newspapers are themselves often metonymically called newspapers.

Newspapers have traditionally been published in print (usually on cheap, low-grade paper called newsprint). However, today most newspapers are also published on websites as online newspapers, and some have even abandoned their print versions entirely.

Newspapers developed in the 17th century, as information sheets for businessmen. By the early 19th century, many cities in Europe, as well as North and South America, published newspapers.

Some newspapers with high editorial independence, high journalism quality, and large circulation are viewed as newspapers of record."""
```

#### Tokenizing the text to words

```python
token_word = word_tokenize(word)
```

```python
token_word
```

```python
from nltk.probability import FreqDist
fdist = FreqDist()
```

#### Finding the frequency of the words occurance

```python
for words in token_word:
    fdist[words.lower()] +=1
fdist    
```

#### Finding the most 3 commonly used words

```python
fdist_top3 = fdist.most_common(3)
fdist_top3
```

#### Finding the number of paragraphs in the text

```python
from nltk.tokenize import blankline_tokenize
word_blank = blankline_tokenize(word)
```

```python
len(word_blank)
```

#### Finding consequtive words. We can find n number of consequtive words from the text:
bigrams:  2 consequtive numbers <br/>
trigrams: 3 consequtive numbers <br/>
ngrams:   n consequtive numbers        

```python
from nltk.util import bigrams, trigrams, ngrams
```

```python
sent = "The best thing in the world cannot be seen or even touched, they must be felt with the heart"
```

```python
sent_token = word_tokenize(sent)
sent_token
```

```python
sent_biagram = list(nltk.bigrams(sent_token))
sent_biagram
```

```python
sent_trigram = list(nltk.trigrams(sent_token))
sent_trigram
```

```python
sent_ngram = list(nltk.ngrams(sent_token, 4))
sent_ngram
```
